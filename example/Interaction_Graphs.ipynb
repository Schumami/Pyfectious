# Interaction Graphs
This notebook is dedicated to draw the graph if interaction between the people in the simulation.
To this end, the 'igraph' library is employed along with a light simulation of 75 people working in predefined work environments.

### Run the Simulation

```python
import sys, os
sys.path.insert(1, os.path.join(os.pardir, 'code'))

# Import and initialize the parser
from json_handle import Parser

parser = Parser(folder_name='graph')

# Load Simulator from JSON file
simulator = parser.parse_simulator()

# Generate the simulation model
simulator.generate_model()

# Load Simulator Data from JSON file
end_time, spread_period, initialized_infected_ids, \
commands, observers = parser.parse_simulator_data()

# Run the simulation
simulator.simulate(end_time=end_time,
                   spread_period=spread_period,
                   initialized_infected_ids=initialized_infected_ids,
                   commands=commands,
                   observers=observers,
                   report_statistics=0)
```

```python
simulator.report_stats(2)
```
